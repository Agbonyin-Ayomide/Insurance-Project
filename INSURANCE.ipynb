```python
insurance["fraud_binary"] = (insurance["fraud_reported"].astype(str).str.upper().str.strip() == "Y").astype(int)
print("\nFraud counts:\n", insurance["fraud_reported"].value_counts())
print("Fraud rate (Y):", insurance["fraud_binary"].mean())
```

```python
# Correlation with fraud (linear relationships only)
num_for_corr = insurance.select_dtypes(include=[np.number]).columns
corr_to_fraud = insurance[num_for_corr].corr(numeric_only=True)["fraud_binary"].drop("fraud_binary").abs().sort_values(ascending=False)
print("\nTop |correlation| with fraud_binary:\n", corr_to_fraud.head(12))
```

```python

# Heatmap of numeric correlations (optional; can be slow if many columns)
plt.figure(figsize=(12, 10))
sns.heatmap(insurance[num_for_corr].corr(), cmap="coolwarm", center=0)
plt.title("Numeric correlation matrix (df1)")
plt.tight_layout()
plt.show()
```

```python
#Key observations from the correlation matrix:age vs months_as_customer (strong red) Older people in this sample tend to have been customers longer. That is a strong positive linear association—often real in insurance data (tenure grows with age on average).

#Claim block: total_claim_amount, injury_claim, property_claim, vehicle_claim (big red block) These move together because they are parts of the same story (amounts that make up or track the overall claim). Very high correlation here often means redundancy: a model might not need all four if one carries almost the same information (or they are definitionally linked).

#fraud_binary row/column (mostly pale) Most correlations with fraud are weak. That means no single numeric column has a strong simple linear relationship with fraud by itself. That is normal for fraud—and it motivates more features, non-linear models, or combinations of variables, not “fraud is explained by one number.”

#Many pale / near-zero cells Lots of pairs (e.g. identifiers or weakly related fields) do not linearly co-move. That is also expected.
```

```python
# Using IQR to identify outliers
Q1 = insurance['total_claim_amount'].quantile(0.25)
Q3 = insurance['total_claim_amount'].quantile(0.75)
IQR = Q3 - Q1

outliers = insurance[(insurance['total_claim_amount'] < (Q1 - 1.5 * IQR)) | (insurance['total_claim_amount'] > (Q3 + 1.5 * IQR))]
print("Number of outliers:", len(outliers))

# Handling outliers (e.g., capping or removing)
insurance['total_claim_amount'] = np.where(insurance['total_claim_amount'] > (Q3 + 1.5 * IQR), (Q3 + 1.5 * IQR), insurance['total_claim_amount'])
```

```python
# Combining a model into a single column
insurance['vehicle'] = insurance['auto_make'] + ' ' + insurance['auto_model']

# Aggregate total claim amount by vehicle
vehicle_claims = insurance.groupby('vehicle')['total_claim_amount'].sum().reset_index()

# Sort the data by total claim amount
vehicle_claims = vehicle_claims.sort_values(by='total_claim_amount', ascending=False)

# Create a horizontal bar chart
plt.figure(figsize=(14, 10))
sns.barplot(x='total_claim_amount', y='vehicle', data=vehicle_claims, palette='viridis', ci=None)
plt.title('Total Claim Amount by Vehicle Make and Model')
plt.xlabel('Total Claim Amount')
plt.ylabel('Vehicle Make and Model')

# Add labels to the bars
for index, value in enumerate(vehicle_claims['total_claim_amount']):
    plt.text(value, index, f'{value:.2f}', color='black', ha="left", va="center")

plt.show()
```

```python
insurance ['vehicle'] = insurance['auto_make'] + ' ' + insurance['auto_model']

# Aggregate total claim amount and fraud count by vehicle
vehicle_fraud = insurance.groupby('vehicle').agg({
    'total_claim_amount': 'sum',
    'fraud_reported': lambda x: (x == 'Y').sum()
}).reset_index()

# Create a scatter plot
plt.figure(figsize=(14, 8))
sns.scatterplot(x='total_claim_amount', y='fraud_reported', data=vehicle_fraud, hue='vehicle', palette='viridis')
plt.title('Total Claim Amount vs. Number of Fraud Cases by Vehicle')
plt.xlabel('Total Claim Amount')
plt.ylabel('Number of Fraud Cases')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
insurance ['vehicle'] = insurance['auto_make'] + ' ' + insurance['auto_model']

# Aggregate total claim amount and fraud count by vehicle
vehicle_fraud = insurance.groupby('vehicle').agg({
    'total_claim_amount': 'sum',
    'fraud_reported': lambda x: (x == 'Y').sum()
}).reset_index()

# Calculate the correlation coefficient
correlation = vehicle_fraud['total_claim_amount'].corr(vehicle_fraud['fraud_reported'])
print("Correlation between total claim amount and number of fraud cases:", correlation)
```

```python

# we should investivate what age group exhibit thee fraudulent claims


# age groups
bins = [18, 25, 35, 45, 55, 65, 75]
labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65-74']
insurance['age_group'] = pd.cut(insurance['age'], bins=bins, labels=labels, right=False)

# fraud count by age group
age_fraud = insurance.groupby('age_group').agg({
    'fraud_reported': lambda x: (x == 'Y').sum()
}).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='age_group', y='fraud_reported', data=age_fraud, palette='viridis')
plt.title('Number of Fraud Cases by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Fraud Cases')
plt.show()

```
